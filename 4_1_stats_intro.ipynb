```python
import pandas as pd
%matplotlib inline
```

```python
obs_ages = pd.read_csv('ages_population.csv')
```

```python
obs_ages
```

```python
obs_ages[['observation']].plot(kind = 'bar', figsize=(50,30))
```

```python
obs_ages[['observation']].sort_values(by='observation').plot(kind = 'bar', figsize=(50,30))
```

```python
freq_df = obs_ages['observation'].value_counts()
```

```python
freq_df = pd.DataFrame(data = freq_df).sort_index().reset_index().rename(columns={'index': 'age'})
```

```python
freq_df
```

```python
freq_df[['observation']].plot(kind='bar', figsize=(50,30))
```

```python
poss_ages = pd.DataFrame(data = {'age': pd.Series(range(100))})
```

```python
poss_ages
```

```python
poss_ages.merge(freq_df, how='left').fillna(0)[['observation']].plot(kind='bar', figsize=(50,30))
```
