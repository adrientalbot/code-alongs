## The whole and cleaned process
Sometimes run a huge process can be very heavy. You can define [chunks](https://en.wikipedia.org/wiki/Chunk_(information)) (little pieces of your dataframes instead of all the dataframe). We are adding that part to the process. 

The Ironhack API allows you to define the start parameter, if you change the variable you will have another starting point in order to have different information (look at the function below).

Note: as the snippet was too long for the number of characters we defined in the db (45 characters) I needed to give more capacity to that field

```python
def call_google_api(q, start, 
                    key="AIzaSyDeDEXWk7bq0r0GYzz-Pmc6dWqRGVxJ2Zc", 
                    cx="003240156206868727193:scrr7eqktoy",
                    fields="kind,items(title,link,snippet),queries(request(title,totalResults,count,startIndex))"):
    params = {"key": key, "cx": cx, "fields": fields, "q": q, "start": start}    
    return requests.get(google_api, params=params).json()
```

```python
def save_city(city):
    for bulk in range(1, 100, 10):
        save_bulk_city(city, bulk)    
```

```python
def save_bulk_city(city, bulk):
    bulk_result = call_google_api(q=city, start=bulk)
    if "items" in bulk_result.keys():
        save_bulk(city, bulk_result)
```

```python
def save_bulk(city, bulk_result):
    for item in bulk_result["items"]:
        save_one_item(conn, city, item)
```

```python
def save_one_item(conn, city, item):
    save_in_city_table(conn, city, item)
    save_in_events_table(conn, city, item)
```

```python
def save_in_city_table(conn, city, item):
    cursor = conn.cursor()
    query = f'INSERT IGNORE INTO cities (name) VALUES ("{city}")'
    cursor.execute(query)
    conn.commit()
    return cursor.rowcount
```

```python
def save_in_events_table(conn, city, item):
    cursor = conn.cursor()
    city_id_query = f'SELECT id FROM cities where name="{city}"'
    city_id = pd.read_sql(city_id_query, conn)["id"].values[0]
    query = "INSERT INTO events (snippet, link, title, city_id) VALUES (%s, %s, %s, %s)"
    cursor.execute(query, (            
        item["snippet"], 
        item["link"],
        item["title"],
        str(city_id)
    ))
    conn.commit()
    return cursor.rowcount
```

```python
def save_all_cities(list_cities, conn):
    for city in list_cities:
        save_city(city)        
```
